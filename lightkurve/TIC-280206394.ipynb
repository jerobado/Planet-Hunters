```python
import lightkurve as lk
result = lk.search_targetpixelfile('tic 280206394')
result
```

```python
tpf = result.download()
lc = tpf.to_lightcurve().remove_outliers(sigma=6)
lc.errorbar()
```

```python
wl = int(round(lc.time.shape[0] * 0.05))
flat_lc = lc.flatten(window_length=wl)
flat_lc.errorbar()
```

```python
scatter_lc = flat_lc.scatter(s=0.1)
```

```python
folded_lc = flat_lc.fold(period=11.236613, t0=1547.5)
folded_lc.errorbar()
```

```python
scatter_lc = folded_lc.scatter(s=0.1)
```

```python
binned_lc = folded_lc.bin(binsize=10)
scatter_lc = binned_lc.scatter(s=0.52)
```

```python
scatter_lc = binned_lc.scatter(s=0.1)
```
